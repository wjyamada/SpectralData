```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
edata = pd.read_csv("DataE.csv")
sdata = pd.read_csv("DataS.csv")
```

```python
x = [float(i) for i in list(edata.columns)[3:]]
y1 = list(edata.loc[edata.ID==25].values[0][3:])
y2 = list(edata.loc[edata.ID==26].values[0][3:])
y3 = list(edata.loc[edata.ID==27].values[0][3:])
y4 = list(edata.loc[edata.ID==28].values[0][3:])
y5 = list(edata.loc[edata.ID==29].values[0][3:])
plt.plot(x,y1,x,y2,x,y3,x,y4,x,y5)
```

```python
x = [float(i) for i in list(sdata.columns)[3:]]
y1 = list(sdata.loc[sdata.ID==25].values[0][3:])
y2 = list(sdata.loc[sdata.ID==26].values[0][3:])
y3 = list(sdata.loc[sdata.ID==27].values[0][3:])
y4 = list(sdata.loc[sdata.ID==28].values[0][3:])
y5 = list(sdata.loc[sdata.ID==29].values[0][3:])
plt.plot(x,y1,x,y2,x,y3,x,y4,x,y5)
```
