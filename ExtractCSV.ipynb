```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
df = pd.read_csv("Data.csv")
df.head()
```

```python
ids = df["ID"].tolist()
```

```python
specs = {}

for i in ids:
    files = glob.glob("TotalScans/"+str(i)+"e*")
    specs[i] ={}
    for file in files:
        with open(file) as fp:
            lines = fp.readlines()
            for line in lines[1:]:
                data = line.replace("\n","").split("\t")
                if data[0] not in specs[i]:
                    specs[i][data[0]] = 0
                specs[i][data[0]] += float(data[1])
```

```python
for i in specs.keys():
    for sp in specs[i].items():
        specs[i][sp[0]] = sp[1]/5
```

```python
di = {}

for _,i in specs.items():
    for j in i.items():
        if j[0] not in di:
            di[j[0]]=[]
        di[j[0]].append(j[1])
```

```python
for d in di.items():
    df[d[0]]=d[1]
```

```python
df = pd.read_csv("Data.csv")
df.head()

ids = df["ID"].tolist()

specs = {}

for i in ids:
    files = glob.glob("TotalScans/"+str(i)+"s*")
    specs[i] ={}
    for file in files:
        with open(file) as fp:
            lines = fp.readlines()
            for line in lines[1:]:
                data = line.replace("\n","").split("\t")
                if data[0] not in specs[i]:
                    specs[i][data[0]] = 0
                specs[i][data[0]] += float(data[1])

for i in specs.keys():
    for sp in specs[i].items():
        specs[i][sp[0]] = sp[1]/5

di = {}

for _,i in specs.items():
    for j in i.items():
        if j[0] not in di:
            di[j[0]]=[]
        di[j[0]].append(j[1])

for d in di.items():
    df[d[0]]=d[1]

df.to_csv("DataS.csv", index=False)
```
